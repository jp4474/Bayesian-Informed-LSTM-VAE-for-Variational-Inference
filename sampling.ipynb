```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
N = 1000
```

```python
alpha = np.random.normal(0.01, 0.5, N)
beta = np.random.normal(0.01, 0.5, N)
mu = np.random.normal(0.01, 0.5, N)
nu = np.random.normal(0.01, 0.5, N)
delta = np.random.normal(0.8, 0.3, N)
lambda_wt = np.random.normal(0.1, 0.3, N)
```

```python
from scipy.stats import truncnorm

def get_truncated_normal(mean=0, sd=1, low=0, upp=10, size = 1):
    return truncnorm(a=(low - mean) / sd, b=(upp - mean) / sd, loc=mean, scale=sd).rvs(size)
```

```python
alpha = get_truncated_normal(mean=0.01, sd=0.5, low=0, upp=1, size=N)
beta = get_truncated_normal(mean=0.01, sd=0.5, low=0, upp=1, size=N)
mu = get_truncated_normal(mean=0.01, sd=0.5, low=0, upp=1, size=N)
nu = get_truncated_normal(mean=0.01, sd=0.5, low=0, upp=1, size=N)
delta = get_truncated_normal(mean=0.8, sd=0.3, low=0, upp=1, size=N)
lambda_wt = get_truncated_normal(mean=0.1, sd=0.3, low=0, upp=1, size=N)
```

```python
df = pd.DataFrame({'alpha': alpha, 'beta': beta, 'mu': mu, 'nu': nu, 'delta': delta, 'lambda_wt': lambda_wt})
```

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
plt.figure(figsize=(15, 10))
plt.subplot(grid[0, 0])
sns.kdeplot(df.alpha, shade=True)
plt.title('alpha')
plt.subplot(grid[0, 1])
sns.kdeplot(df.beta, shade=True)
plt.title('beta')
plt.subplot(grid[0, 2])
sns.kdeplot(df.mu, shade=True)
plt.title('mu')
plt.subplot(grid[1, 0])
sns.kdeplot(df.nu, shade=True)
plt.title('nu')
plt.subplot(grid[1, 1])
sns.kdeplot(df.delta, shade=True)
plt.title('delta')
plt.subplot(grid[1, 2])
sns.kdeplot(df.lambda_wt, shade=True)
plt.title('lambda_wt')
plt.show()
```
